# Afghanistan

### Importing Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Basic Stuff

```python
df = pd.read_csv('Life_Expectancy_Data.csv')
```

```python
df.head()
```

```python
df.info()
```

## Selecting Afghanistan as 1st country to test

```python
Afghanistan = df[df['Country'] == 'Afghanistan']
```

```python
Afghanistan
```

## Select Multiple columns for Multi-D model

```python
x = Afghanistan[['Year','Adult Mortality','infant deaths','Total expenditure', 'GDP']]
y = Afghanistan.iloc[:, 3].values # columns 3 = Life expectancy
```

```python
x
```

```python
y
```

## Training the data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

## Testing the accuaracy of our model

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Caculating R^2 of our model

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

### Prediction of life expectancy in the year 2100

```python
Year = 2100
Adult_Mortality = Afghanistan['Adult Mortality'].mean()
infant_death = Afghanistan['infant deaths'].mean()
total_expenditure = Afghanistan['Total expenditure'].mean()
GDP = Afghanistan['GDP'].mean()
```

```python
regressor.predict([[Year,Adult_Mortality,infant_death,total_expenditure,GDP]])
```
